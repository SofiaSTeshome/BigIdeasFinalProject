# The Most Livable Neighborhood in Pittsburgh Over Time

The first metric we used to determine the most livable neighborhood was the air quality. The dataset that I analyzed contained the daily air quality index for each neighborhood from 2016-2025.

```python
import pandas as pd
data = pd.read_csv("DailyAQI.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

I first grouped the data to filter out everything except the numerical AQI for each neighborhood. I then calculated the average AQI for each neighborhood. 

```python
grouped=data.groupby('site')['index_value']
mean=grouped.mean()
```

Two of the neighborhoods were duplicates, so I removed them from the dataset to make it cleaner. Then, I printed out the average AQI for all of the neighborhoods.

```python
fixed=mean.drop(['Lawrenceville 2', 'Liberty 2'])
print(fixed)
```

Finally, I created a bar graph to visualize the data. I also sorted the data from lowest to highest AQI so that we could clearly see how the neighborhoods ranked.

```python
fixed.sort_values().plot(kind='bar')
```

Since a lower AQI means cleaner air, we can see that West Mifflin has the best air quality out of all the neighborhoods.

Our second metric was the population of each neighborhood, which was taken from 2010 and 2020 census. We decided that the neighborhoods with a higher population were more livable because a high population means that a lot of people want to live in that area.

```python
data = pd.read_csv("Population.csv")
```

I first isolated the most recent data, which was from 2020. I then printed out the average population for the 10 neighborhoods with the highest population.

```python
grouped=data.groupby('Neighborhood')['2020_Total_Population']
mean=grouped.mean().sort_values(ascending=False).head(10)
print(mean)
```

Lastly, I created a bar graph to visualize the data. I also sorted the values again to make the graph more readable.

```python
mean.sort_values().plot(kind='bar')
```

We can see from the graph that Squirrel Hill South and Shadyside are tied for the highest population.

The third dataset that we used the was crime rate from 1973-1975. We wanted to see which neighborhoods had the lowest crime rate throughout those years to use for our claim.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
data = pd.read_csv("CrimeRate1977.csv")
```

To get the best visual of the data, I chose to group each catagory together from each year to get a better visual of the individual categories.



```python
groups=data.groupby('neighborhood')[['number_major_crimes_1973', 'neighborhood_crime_rate_1973', 'number_major_crimes_1974', 'neighborhood_crime_rate_1974', 'number_major_crimes_1975', 'neighborhood_crime_rate_1975']]
groups.mean()
```

After I took a look at those numbers, I wanted to visualize how the average crime rates changed over those years. I could then see how the crime rates increased. When looking at the data above. We can see that from those neighborhoods with an extremely high crime rate like Uptown and East Liberty, the affect the graph significantly taking away the change from the other neighborhoods that had a decrease over time. Although the majority of neighborhoods seemed to have a slight decline.

```python
means = groups.mean()
years = [1973, 1974, 1975]
crime_rates = means[['neighborhood_crime_rate_1973', 'neighborhood_crime_rate_1974', 'neighborhood_crime_rate_1975']].mean()

plt.plot(years, crime_rates, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Crime Rate')
plt.title('Average Crime Rate Over Time')
plt.grid(True)
plt.show()
```

I then wanted to further dive deep into my assumption by finding the average crime rates from 1973 and 1975 and to get a better visual.

```python
grouped=data.groupby('neighborhood')['neighborhood_crime_rate_1973']
mean=grouped.mean()
print(mean)
```

```python
grouped=data.groupby('neighborhood')['neighborhood_crime_rate_1975']
mean=grouped.mean()
print(mean)
```

Overall, it seemed that Glen Hazel Heights had the lowest crime rate, including the number of major crimes.
